```python
def columnEncoder(column):
    values = column
    values = values[~pd.isna(values)].unique()
    Dict = dict(zip(values, np.arange(1, values.shape[0] + 1, 1, dtype=np.int64)))
    return pd.to_numeric(column.replace(Dict), downcast='integer'), Dict
```

```python
previousDF["rating"], ratingDict = columnEncoder(previousDF["rating"])
previousDF["genre"], genreDict = columnEncoder(previousDF["genre"])
previousDF["company"], companyDict = columnEncoder(previousDF["company"])
```

```python
def getInforamtionOfcolumns(df: pd.DataFrame, groupbyList: list, columns: list, aggFunction: list) -> pd.DataFrame:
    myDict = {}
    for i in columns:
        myDict[i] = aggFunction

    return df.groupby(groupbyList).agg(myDict)
```

```python
ratingInformation = getInforamtionOfcolumns(df, ["rating"], ["gross", "budget", "runtime", "score"], ["mean", "max", "min", "std"])
ratingInformation
```

```python
# reverse dictionary

# solution 1 :

# temp = {}
# for i, j in ratingDict.items():
#     temp[j] = i

# solution 2 :
# {j: i for i, j in ratingDict.items()}

# result :
# {
    #  1: 'R',
    #  2: 'PG',
    #  3: 'G',
    #  4: 'Not Rated',
    #  5: 'NC-17',
    #  6: 'Approved',
    #  7: 'TV-PG',
    #  8: 'PG-13',
    #  9: 'Unrated',
    #  10: 'X',
    #  11: 'TV-MA',
    #  12: 'TV-14'
# }
```

```python
def getInformation(df: pd.DataFrame, columns: list, info: str, myDict: dict):
    # revers dictioanry
    myDict = {v: k for k, v in myDict.items()}
    if (info == "mean"):
        for i in columns:
            print(f"Max mean of [{i}] -> {int(df[i]['mean'].idxmax())} [{myDict[int(df[i]['mean'].idxmax())]}]")
            print(f"Min mean of [{i}] -> {int(df[i]['mean'].idxmin())} [{myDict[int(df[i]['mean'].idxmin())]}]")
            print()
    elif (info == "std"):
        for i in columns:
            print(f"Max mean of [{i}] -> {int(df[i]['std'].idxmax())} [{myDict[int(df[i]['std'].idxmax())]}]")
            print(f"Min mean of [{i}] -> {int(df[i]['std'].idxmin())} [{myDict[int(df[i]['std'].idxmin())]}]")
            print()
    else:
        pass
```

```python
getInformation(ratingInformation, ["gross", "budget", "runtime", "score"], "mean", ratingDict)
```

```python
getInformation(ratingInformation, ["gross", "budget", "runtime", "score"], "std", ratingDict)
```

```python
def myFillNA(df: pd.DataFrame, discreteColumns: list, continuousColumns: list) -> pd.DataFrame:
    for col in continuousColumns:
        df.loc[::, col] = df.loc[::, col].fillna(df.loc[::, col].mean())
    
    for col in discreteColumns:
        df.loc[::, col] = df.loc[::, col].fillna(df.loc[::, col].mode()[0])

    return df
```

```python
discrete = ["rating", "genre", "company"]
continuous = [elmt for elmt in df.columns if elmt not in discrete]
```

```python
newDF = myFillNA(df, discrete, continuous)
```

```python
newDF.isna().sum()
```

```python
corrDF = newDF.corr()
corrDF
```

```python
# get upper triangle of correlation matrix
upper = corrDF.where(np.triu(np.ones(corrDF.shape), k=1).astype(bool))
upper
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 9))
cax = ax.matshow(upper, cmap="Reds") # cmap="rainbow"
ax.set_xticklabels(['']+list(df.columns))
ax.set_yticklabels(['']+list(df.columns))
fig.colorbar(cax);
```

```python
positiveCorr = upper[upper > 0]
positiveCorr
```

```python
negetiveCorr = upper[upper < 0]
negetiveCorr
```

```python
neutralCorr = upper[(upper > 0) & (upper < 0.2)]
neutralCorr
```
